```
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

#setup
lines_num = 50

def decel(x):
  return 1-(x-1)*(x-1)

uploaded = files.upload()
for k, v in uploaded.items():
    inp_img = plt.imread(k)
plt.imshow(inp_img)
plt.show()

inp_shape = inp_img.shape
out_img = np.ones(inp_shape)
for line in range(0, lines_num):
  l=0
  y = int(line*inp_shape[0]/lines_num)
  for x in range(0, inp_shape[1]):
    m = 1 - np.amin(np.max(inp_img[y, x]))      
    for s in range(0, 6):
      l+=m/4
      out_img[y + int(np.sin(l * np.pi/2.) * 5 * decel(m)), x] = 0

plt.imshow(out_img)
plt.imsave('{}_lines_{}'.format(lines_num, k),out_img)
plt.show()
```
